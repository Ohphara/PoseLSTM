```python
import os
import glob
import cv2
import json
import pandas as pd
from tqdm import tqdm
from ultralytics import YOLO
import logging
from ultralytics.utils import LOGGER

LOGGER.setLevel(logging.WARNING)

# 모델 로드
model = YOLO("yolo11n-pose.pt")  # 공식 pretrained 모델

# 처리할 폴더 목록
folders = [
    # 예시 폴더 경로들
    #'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/01.원천데이터/TS_03.이상행동_07.전도',
    #'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/01.원천데이터/TS_03.이상행동_08.파손',
    ##'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/01.원천데이터/TS_03.이상행동_09.방화',
    ##'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/01.원천데이터/TS_03.이상행동_10.흡연',
    ##'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/01.원천데이터/TS_03.이상행동_11.유기',
    #'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/01.원천데이터/TS_03.이상행동_12.절도',
    #'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/01.원천데이터/TS_03.이상행동_13.폭행',
    #'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/01.원천데이터/TS_03.이상행동_14.교통약자',

    #'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS_03.이상행동_07.전도',
    #'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS_03.이상행동_08.파손',
    ##'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS_03.이상행동_09.방화',
    ##'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS_03.이상행동_10.흡연',
    ##'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS_03.이상행동_11.유기',
    #'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS_03.이상행동_12.절도',
    #'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS_03.이상행동_13.폭행',
    #'/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS_03.이상행동_14.교통약자',
    # 필요에 따라 추가
]

# 각 폴더 내의 .mp4 파일을 순회
for folder in folders:
    mp4_files = glob.glob(os.path.join(folder, '*.mp4'))
    for video_path in tqdm(mp4_files, desc=f"Processing videos in {folder}"):
        # 결과 저장 경로 설정
        base_name = os.path.splitext(os.path.basename(video_path))[0]
        # output_json = os.path.join(folder, f"{base_name}_pose.json")
        output_csv = os.path.join(folder, f"{base_name}_pose.csv")

        # 비디오 캡처 객체 생성
        cap = cv2.VideoCapture(video_path)
        frame_idx = 0
        results_list = []

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            # 모델 추론
            results = model.predict(source=frame, save=False, stream=False)

            # 결과에서 키포인트 추출
            for result in results:
                if result.keypoints is not None:
                    keypoints = result.keypoints.xy.cpu().numpy().tolist()  # [num_persons, num_keypoints, 2]
                    for person_idx, person_keypoints in enumerate(keypoints):
                        result_dict = {
                            "frame": frame_idx,
                            "person_id": person_idx,
                            "keypoints": person_keypoints
                        }
                        results_list.append(result_dict)

            frame_idx += 1

        cap.release()

        # # 결과를 JSON으로 저장
        # with open(output_json, 'w', encoding='utf-8') as f:
        #     json.dump(results_list, f, ensure_ascii=False, indent=4)

        # 결과를 CSV로 저장
        csv_rows = []
        for item in results_list:
            row = {
                "frame": item["frame"],
                "person_id": item["person_id"]
            }
            for idx, (x, y) in enumerate(item["keypoints"]):
                row[f"x{idx}"] = x
                row[f"y{idx}"] = y
            csv_rows.append(row)

        df = pd.DataFrame(csv_rows)
        df.to_csv(output_csv, index=False)

        print(f"Processed {video_path}. Results saved to {output_csv}")
```

## - 라벨링데이터에서 모델 훈련과 평가에 필요한 라벨링만 추출하여 csv파일로 저장

```python
import xml.etree.ElementTree as ET
import pandas as pd
from pathlib import Path

class_mapping = {
    'fall': 7,
    'broken': 8,
    'theft': 12,
    'fight': 13,
    'weak pedestrian': 14
}

def parse_xml_to_csv(xml_path, csv_output_path, total_frames=180):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    labels = [0] * total_frames

    events = {}

    for track in root.findall('track'):
        label_name = track.get('label')
        action = label_name.split('_')[0]
        for box in track.findall('box'):
            frame = int(box.get('frame'))
            outside = box.get('outside')

            if outside == '0':
                if 'start' in label_name:
                    events.setdefault(action, {})['start'] = frame
                elif 'end' in label_name:
                    events.setdefault(action, {})['end'] = frame

    for action, frames in events.items():
        class_num = class_mapping.get(action, 0)
        start_frame = frames.get('start', 0)
        end_frame = frames.get('end', total_frames - 1)

        for i in range(start_frame, end_frame + 1):
            if 0 <= i < total_frames:
                labels[i] = class_num

    df = pd.DataFrame({
        'frame': list(range(total_frames)),
        'label': labels
    })

    df.to_csv(csv_output_path, index=False)
    print(f'Labels saved to {csv_output_path}')


# 라벨링데이터 경로
labeling_folders = [
    '/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/02.라벨링데이터/TL_03.이상행동_07.전도',
    '/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/02.라벨링데이터/TL_03.이상행동_08.파손',
    '/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/02.라벨링데이터/TL_03.이상행동_12.절도',
    '/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/02.라벨링데이터/TL_03.이상행동_13.폭행',
    '/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Training/02.라벨링데이터/TL_03.이상행동_14.교통약자',

    '/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/02.라벨링데이터/VL_03.이상행동_07.전도',
    '/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/02.라벨링데이터/VL_03.이상행동_08.파손',
    '/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/02.라벨링데이터/VL_03.이상행동_12.절도',
    '/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/02.라벨링데이터/VL_03.이상행동_13.폭행',
    '/content/drive/MyDrive/25_IAP/238-2.실내(편의점, 매장) 사람 이상행동 데이터/01-1.정식개방데이터/Validation/02.라벨링데이터/VL_03.이상행동_14.교통약자'
]

# 각 폴더 내 모든 xml 파일 파싱
for folder in labeling_folders:
    xml_files = sorted(Path(folder).glob('*.xml'))
    for xml_file in xml_files:
        csv_output_path = xml_file.with_suffix('.csv')
        parse_xml_to_csv(xml_file, csv_output_path)
```
